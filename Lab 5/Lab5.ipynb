## **Q1.** Take two sorted arrays as input, Write a program to merge them into a new array in such a way that the resultant array is sorted.

```python
n1 = int(input("Enter the length of the first array: "))
arr1=[]
for i in range(n1):
    arr1.append(int(input("Enter the element: ")))

n2 = int(input("Enter the length of the second array: "))
arr2=[]
for i in range(n2):
    arr2.append(int(input("Enter the element: ")))

print("The first array is: ",arr1)
print("The second array is: ",arr2)
print("The merged array is: ")

# Here the array is initialized with the size of n1 and n2
# [0] * (n1 + n2) indicates that the array(arr3) is declared of the size n1 and n2
arr3 = [0] * (n1 + n2)

i=0
j=0
k=0

while i < n1 and j < n2:
    if arr1[i] < arr2[j]:
        arr3[k] = arr1[i]
        k = k + 1
        i = i + 1
    else:
        arr3[k] = arr2[j]
        k = k + 1
        j = j + 1

while i < n1:
    arr3[k] = arr1[i]
    k = k + 1
    i = i + 1
    
while j < n2:
    arr3[k] = arr2[j]
    k = k + 1
    j = j + 1

for i in range(len(arr3)):
    print(arr3[i], end = " ")
```

# **Q4.** Do the Q3 in Single pass i.e O(n).

```python
arr = [12, 11, 13, 5, 6]

for i in range(1, len(arr)):
	key = arr[i]
	j = i-1
	while j >=0 and key < arr[j] :
		arr[j+1] = arr[j]
		j -= 1
	arr[j+1] = key

print ("Sorted array is:")

for i in range(len(arr)):
	print ("%d" %arr[i])
```

# **Q5.** Calculate number of comparisons between elements made and swapping of elements made in above programs.

```python
arr = [12, 11, 13, 5, 6]
comp = 0
swap = 0
for i in range(1, len(arr)):
	key = arr[i]
	j = i-1
	while j >=0 and key < arr[j] :
		comp+=1
		arr[j+1] = arr[j]
		j -= 1
	arr[j+1] = key
	swap+=1

print("The number of comparasions made is: ",comp)
print("The number of swaps made is: ",swap)

print ("Sorted array is:")

for i in range(len(arr)):
	print ("%d" %arr[i])
```
