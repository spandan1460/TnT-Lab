```python
import time
#selection sort
def selection_sort(data,size,c):
  selection_time=time.time()
  for i in range(0,size):
    min=i
    for j in range(i+1,size):
      if data[j]<data[min]:
        min=j
    data[i]=data[min]
    data[min]=data[i]
  if c==1:
    print("The sorted dataset is:")
    for i in range(0,size):
      print(data[i])
  selection_time=time.time() - selection_time
  return selection_time

#insertion sort
def insertion_sort(data,size,c):
  insertion_time=time.time()
  for i in range(1,size):
    key=data[i]
    j=i-1
    while j>=0 and key<data[j]:
      data[j+1]=data[j]
      j=j-1
  data[j+1]=key
  if c==2:
    print("The sorted dataset is:")
  for i in range(0,size):
     print(data[i])
  insertion_time=time.time() - insertion_time
  return insertion_time

#bubble sort
def bubble_sort(data,size,c):
  bubble_time=time.time()
  for i in range(0,size):
    for j in range(0,size-i-1):
      if data[j]>data[j+1]:
        temp=data[j]
        data[j]=data[j+1]
        data[j+1]=temp
  if c==3:
    print("The sorted dataset is:")
    for i in range(0,size):
      print(data[i])
  bubble_time=time.time() - bubble_time
  return bubble_time

#Quick sort
def quick_sort(data, low, high,size,c):
  pivot = data[high]
  i = low - 1
  for j in range(low, high):
    if data[j] <= pivot:
      i = i + 1
      (data[i], data[j]) = (data[j], data[i])
  (data[i + 1], data[high]) = (data[high], data[i + 1])
  return i + 1
def quickSort(data, low, high,c):
  quick_time=time.time()
  if low < high:
    pi = quick_sort(data, low, high,c)
    quickSort(data, low, pi - 1,c)
    quickSort(data, pi + 1, high,c)
  if c==4:
    print("The sorted dataset is:")
    for i in range(0,len(data)):
      print(data[i])
  quick_time=time.time() - quick_time
  return quick_time

#merge sort
def merge_sort(data,c):
    merge_time=time.time()
    if len(data) > 1:
        r = len(data)//2
        L = data[:r]
        M = data[r:]
        merge_sort(L,c)
        merge_sort(M,c)
        i = j = k = 0
        while i < len(L) and j < len(M):
            if L[i] < M[j]:
                data[k] = L[i]
                i += 1
            else:
                data[k] = M[j]
                j += 1
            k += 1
        while i < len(L):
            data[k] = L[i]
            i += 1
            k += 1
        while j < len(M):
            data[k] = M[j]
            j += 1
            k += 1
    if c==4:
        print("The sorted dataset is:")
        for i in range(0,len(data)):
           print(data[i])
    merge_time=time.time() - merge_time
    return merge_time

n=int(input("Enter the size of dataset:"))
print("Enter the elements of dataset:")
dataset=[]
for i in range(0,n):
  element=int(input())
  dataset.append(element)
print("Enter 1 for selection sort ")
print("Enter 2 for Insertion sort ")
print("Enter 3 for bubble sort ")
print("Enter 4 for Quick sort ")
print("Enter 5 for merge sort ")
ch=int(input())
if ch==1:
  selection_sort(dataset,n,ch)
if ch==2:
  insertion_sort(dataset,n,ch)
if ch==3:
  bubble_sort(dataset,n,ch)
if ch==4:
  quick_sort(dataset,0,n-1,n,ch)
if ch==5:
  merge_sort(dataset,ch)
ch=0
s_time=selection_sort(dataset,n,ch)
i_time= insertion_sort(dataset,n,ch)
b_time=bubble_sort(dataset,n,ch)
q_time=quick_sort(dataset,0,n-1,n,ch)
m_time=merge_sort(dataset,ch)
if s_time<i_time and s_time<b_time and s_time<q_time and s_time<m_time:
  print("Selection sort is the best")
elif i_time<s_time and i_time<b_time and i_time<q_time and i_time<m_time:
  print("Insertion sort is the best")
elif b_time<s_time and b_time<i_time and b_time<q_time and b_time<m_time:
  print("bubble sort is the best")
elif q_time<s_time and q_time<b_time and q_time<i_time and q_time<m_time:
  print("quick sort is the best")
elif m_time<s_time and m_time<b_time and m_time<i_time and m_time<q_time:
  print("merge sort is the best")
```

```python
class Solution:
    def solve(self, arr):
        sort_seq = sorted(arr)
        table = {}

        for i, n in enumerate(arr):
            table[n] = i

        swaps = 0
        for i in range(len(arr)):
            n = arr[i]
            s_n = sort_seq[i]
            s_i = table[s_n]

            if s_n != n:
                swaps += 1
                arr[s_i] = n
                arr[i] = s_n
                table[n] = s_i
                table[s_n] = i

        return swaps

ob = Solution()


n= int(input("Enter the number of elements: "))
arr=[]
for i in range(n):
    l=int(input())
    arr.append(l)

print ("The input array is: ",arr)
swaps = ob.solve(arr)
if(swaps == 0):
    print("Array is already sorted")
elif swaps == ((n-1)/2):
    print("Array is in reverse order")
else:
    print("Array is randomly sorted")


print("Number of swaps needed to sort it: ",swaps)
print("The sorted array is: ",arr)
```

```python
def identify_data(data):
    """
    Identify if the data is already sorted, reverse sorted or random
    :param data: list of integers
    :return: string
    """
    if data == sorted(data):
        return "sorted"
    elif data == sorted(data, reverse=True):
        return "reverse sorted"
    else:
        return "random"
```

```python
def bubble_sort_order(data, p):
    """
    Bubble Sort list in ascending order if p=1 and descending order if p=-1
    :param data: list of integers
    :param p: 1 or -1
    :return: tuple(list, float) 
    """
    data_copy = data.copy()
    swaps = 0
    comp = 0
    for i in range(len(data_copy)):
        swapped = False
        for j in range(len(data_copy)-1):
            if p == 1 and data_copy[j] > data_copy[j+1]:
                data_copy[j], data_copy[j+1] = data_copy[j+1], data_copy[j]
                swapped = True
                swaps += 1
                comp += 1
            elif p == -1 and data_copy[j] < data_copy[j+1]:
                data_copy[j], data_copy[j+1] = data_copy[j+1], data_copy[j]
                swapped = True
                swaps += 1
                comp += 1
        if swapped == False:
            break
    time = (comp + 3*swaps) * 10**-6
    return data_copy, time
```

```python
def selection_sort_order(data, p):
    """
    Selection Sort list in ascending order if p=1 and descending order if p=-1
    :param data: list of integers
    :param p: 1 or -1
    :return: tuple(list, float)
    """
    data_copy = data.copy()
    swaps = 0
    comp = 0
    for i in range(len(data_copy)):
        min_index = i
        for j in range(i+1, len(data_copy)):
            if p == 1 and data_copy[j] < data_copy[min_index]:
                min_index = j
                comp += 1
            elif p == -1 and data_copy[j] > data_copy[min_index]:
                min_index = j
                comp += 1
        data_copy[i], data_copy[min_index] = data_copy[min_index], data_copy[i]
        swaps += 1
    time = (comp + 3*swaps) * 10**-6
    return data_copy, time
```

```python
def insertion_sort_order(data, p):
    """
    Insertion Sort list in descending order
    :param data: list of integers
    :param p: 1 or -1
    :return: tuple(list, float)
    """
    data_copy = data.copy()
    swaps = 0
    comp = 0
    for i in range(1, len(data_copy)):
        key = data_copy[i]
        j = i-1
        while j >= 0 and (p == 1 and key < data_copy[j] or p == -1 and key > data_copy[j]):
            data_copy[j + 1] = data_copy[j]
            j -= 1
            data_copy[j + 1] = key
            swaps += 1
            comp += 1
    time = (comp + 3*swaps) * 10**-6
    return data_copy, time
```

```python
def quicksort(nums, lower, upper):
    swaps = 0
    comp = 0
    if lower < upper:
        pivot, swaps, comp = partition(nums, lower, upper, swaps, comp)
        quicksort(nums, lower, pivot - 1)
        quicksort(nums, pivot + 1, upper)
    
    time = (comp + 3*swaps) * 10**-6
    return nums, time

def partition(nums, lower, upper, swaps, comp):
    
    pivot = nums[upper]
    
    i = lower
    
    for j in range(lower, upper):
        if nums[j] < pivot:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1
            swaps += 1
            comp += 1
            
    nums[i], nums[upper] = nums[upper], nums[i]
    swaps += 1
    
    return i, swaps, comp

def quick_sort_order(data, p):
    """
    Quick Sort list in ascending order if p=1 and descending order if p=-1
    :param data: list of integers
    :param p: 1 or -1
    :return: tuple(list, float)
    """
    data_copy = data.copy()
    data_copy, time = quicksort(data_copy, 0, len(data_copy)-1)
    if p == -1:
        data_copy.reverse()
    return data_copy, time
```

```python
def merge_sort_order(data, p):
    """
    Merge Sort list in ascending order if p=1 and descending order if p=-1
    :param data: list of integers
    :param p: 1 or -1
    :return: list
    """
    data_copy = data.copy()
    comp = 0
    swaps = 0
    if len(data_copy)>1:
        intMidValue = len(data_copy)//2
        listLeftHalf = data_copy[:intMidValue]
        listRightHalf = data_copy[intMidValue:]

        merge_sort_order(listLeftHalf, p)
        merge_sort_order(listRightHalf, p)

        i=0
        j=0
        k=0
        while i < len(listLeftHalf) and j < len(listRightHalf):
            if p == 1 and listLeftHalf[i] < listRightHalf[j]:
                data_copy[k]=listLeftHalf[i]
                i=i+1
                comp += 1
                swaps += 1
            elif p == -1 and listLeftHalf[i] > listRightHalf[j]:
                data_copy[k]=listLeftHalf[i]
                i=i+1
                comp += 1
                swaps += 1
            else:
                data_copy[k]=listRightHalf[j]
                j=j+1
                comp += 1
                swaps += 1

            k=k+1

        while i < len(listLeftHalf):
            data_copy[k]=listLeftHalf[i]
            i=i+1
            k=k+1
            comp += 1
            swaps += 1

        while j < len(listRightHalf):
            data_copy[k]=listRightHalf[j]
            j=j+1
            k=k+1
            comp += 1
            swaps += 1

    time = (comp + 3*swaps) * 10**-6
    return data_copy, time
```

```python
def SORT_COMPARE(n, p):
    """
    Sorts an array of n random integers using the 5 sorting algorithms
    and prints the time taken to sort each array.
    """
    # Generate an array of n random integers
    array = [randint(0, 100) for _ in range(n)]
    print("Actual Array:", array)
    data_type = identify_data(array)
    if data_type == "sorted":
        print("Array is already Sorted:", array)
    elif data_type == "reversed":
        print("Array is already Reversed:", array)
    else:

    
      start_time = time.time()
      sorted, time_req = bubble_sort_order(array, p=p)
      stop_time = time.time()
      print(f"\nBubble Sort: {sorted} \n"
            f"Actual Time Consumed by Code: {stop_time - start_time} seconds \n"
            f"Time Required to Sort: {time_req} seconds")
      
      start_time = time.time()
      sorted, time_req = selection_sort_order(array, p=p)
      stop_time = time.time()
      print(f"\nSelection Sort: {sorted} \n"
            f"Actual Time Consumed by Code: {stop_time - start_time} seconds \n"
            f"Time Required to Sort: {time_req} seconds")
      
      start_time = time.time()
      sorted, time_req = insertion_sort_order(array, p=p)
      stop_time = time.time()
      print(f"\nInsertion Sort: {sorted} \n"
            f"Actual Time Consumed by Code: {stop_time - start_time} seconds \n"
            f"Time Required to Sort: {time_req} seconds")
      
      start_time = time.time()
      sorted, time_req = quick_sort_order(array, p=p)
      stop_time = time.time()
      print(f"\nQuick Sort: {sorted} \n"
            f"Actual Time Consumed by Code: {stop_time - start_time} seconds \n"
            f"Time Required to Sort: {time_req} seconds")
      
      start_time = time.time()
      sorted, time_req = merge_sort_order(array, p=p)
      stop_time = time.time()
      print(f"\nMerge Sort: {sorted} \n"
            f"Actual Time Consumed by Code: {stop_time - start_time} seconds \n"
            f"Time Required to Sort: {time_req} seconds")
```

```python
print("Sorting a random array of 10 elements in ascending order:")
SORT_COMPARE(10, 1)

print("Sorting a random array of 10 elements in descending order:")
SORT_COMPARE(10, -1)
```
