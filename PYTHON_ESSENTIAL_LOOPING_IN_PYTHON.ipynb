# Write a program to Find odd numbers and even numbers from a range of number 

```python
n = int(input("Enter the Number"))
i = 0
while i < n:
    if(i%2 == 0):
        print(i, " is even")
    else:
        print(i," is odd")
    i = i + 1
        
```

# break STATEMENT

break is used to break the loop. When the break is encountered the control went out of the loop and next line is executed after the loop.

```python
x = 10
i = 0
while True:
    print('i = ',i, 'Hence Loop continues')
    i += 1
    if i >= 10:
        break
print('break is encountered, hence control came out of the loop')
```

```python
i = 0
while True:
  x = int(input('Enter your Choice, Enter 0 to terminate'))
  if x == 0:
    print('Thank You')
    break
    print('Bye Bye')
  print('You Have entere ', x)
print('You Have come out of the loop')
```

# continue STATEMENT

Upon encoutering continue keyword all the statement in the block in which is key word is excluded from execution and the control came out of the block. 

```python
x = 10
i = -1
while i <= 10 :
    print('The Loop begins again')
    i += 1
    if i % 2 == 0:
        print('i % 2 = ',i%2,'Continue executed', 'Hi Skipped')
        continue
        print('Hi')
    print('i % 2 = ',i%2, 'Hence Loop continues')
#print('break is encountered, hence control came out of the loop')
```
